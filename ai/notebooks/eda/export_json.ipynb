```python
import duckdb
import json
import os
```

```python
db_path = "../../data/data.duckdb"
analytics_path = "../../data/analytics.duckdb"
```

```python
conn = duckdb.connect(analytics_path)
```

```python
tables = conn.execute("""
    SELECT table_name 
    FROM duckdb_tables() 
    WHERE schema_name = 'main';
""").fetchall()
```

```python
tables = [t[0] for t in tables]
```

```python
outdir = "../../../app/data/analytics"
```

```python
for table in tables:
    print(f"Exporting table: {table}")
    # Fetch all rows as a list of dicts
    rows = conn.execute(f"SELECT * FROM {table}").fetchall()
    columns = [desc[0] for desc in conn.description]  # get column names
    data = [dict(zip(columns, row)) for row in rows]

    # Write JSON file
    output_file = os.path.join(outdir, f"{table}.json")
    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=2)

print("All tables exported to JSON!")
```

```python
for table in tables:
    print(f"Exporting table: {table}")
    rows = conn.execute(f"SELECT * FROM {table}").fetchall()
    columns = [desc[0] for desc in conn.description]  # get column names
    data = [dict(zip(columns, row)) for row in rows]

    # Write JS module
    output_file = os.path.join(outdir, f"{table}.js")
    with open(output_file, "w", encoding="utf-8") as f:
        f.write(f"export const {table} = ")
        f.write(str(data).replace("'", '"'))  # convert to JSON-style string
        f.write(";\n")

print("All analytics tables exported as JS modules!")
```

```python
conn.close()
```
