```python
import pandas as pd
import duckdb
```

```python
path = "../../data/dataset.csv"
df = pd.read_csv(path) 
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
db_path = "../../data/data.duckdb"
conn = duckdb.connect(db_path)
```

```python
conn.register("Attention_Span", df)
```

```python
conn.execute("""
    CREATE OR REPLACE TABLE span_data AS 
    SELECT * FROM Attention_Span
""")
```

```python
conn.execute("DESCRIBE span_data")
```

```python
description = conn.execute("DESCRIBE span_data").df()
print(description)
```

```python
def query(param: str) -> None:
    result = conn.execute(param).df()
    print(result)
```

```python
query("""
SELECT * FROM span_data
LIMIT 5;
""")
```

```python
query("""
SELECT Reason_of_low_attention_span, COUNT(*) AS count
FROM span_data
GROUP BY Reason_of_low_attention_span;
""")
```

```python
conn.close()
```
